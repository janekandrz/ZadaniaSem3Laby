```python
import pandas as pd 
import matplotlib.pyplot as plt 


df = pd.read_csv("laby9.csv")

plt.figure(figsize=(7,6))
plt.plot(df["Vwe"],df["Vwy"])
plt.xlabel("Vwe")
plt.ylabel("Vwy")
plt.show()
```

```python
df1 = pd.read_csv("laby91.csv")
df2 = pd.read_csv("laby92.csv")

plt.figure(figsize=(7,6))
plt.plot(df1["Vwe"], df1["Vwy"], label="0 do 12V")
plt.plot(df2["Vwe"], df2["Vwy"], label="12 do 0V")
plt.legend()
plt.title("Wykres histerezy")
plt.xlabel("Vwe")
plt.ylabel("Vwy")
plt.show()
```
